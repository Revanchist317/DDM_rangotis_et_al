```python
# Load all modules 
import hddm
import multiprocessing as mp

# Load data
humans_cyl_str_eng = hddm.load_csv('cyl_humans_glmhmm_for_ddm (1).csv')
humans_rdk_eng = hddm.load_csv('RDK_eng_humans_glmhmm_for_ddm.csv')
humans_rdk_eng_wbias = hddm.load_csv('RDK_eng_w_bias_humans_glmhmm_for_ddm.csv')
```

```python
humans_cyl_str_eng
```

```python
humans_cyl_str_eng.set_axis(["subj_idx", "stim", "rt", "response", "accuracy", "state"], axis=1, inplace=True)
humans_rdk_eng.set_axis(["subj_idx", "stim", "rt", "response", "accuracy", "state"], axis=1, inplace=True)
humans_rdk_eng_wbias.set_axis(["subj_idx", "stim", "rt", "response", "accuracy", "state"], axis=1, inplace=True)
```

```python
humans_cyl_str_eng.drop(columns="state", inplace=True)
humans_rdk_eng.drop(columns="state", inplace=True)
humans_rdk_eng_wbias.drop(columns="state", inplace=True)
```

```python
humans_cyl_str_eng["subj_idx"] = humans_cyl_str_eng["subj_idx"].astype(int)
humans_rdk_eng["subj_idx"] = humans_cyl_str_eng["subj_idx"].astype(int)
humans_rdk_eng_wbias["subj_idx"] = humans_cyl_str_eng["subj_idx"].astype(int)
```

```python
v_reg = {'model': 'v ~ stim', 'link_func': lambda x: x}
```

```python
label = "GLM_HMM_Cyl_H" 
model = hddm.HDDMRegressor(humans_cyl_str_eng, v_reg, group_only_regressors=False, 
                               keep_regressor_trace=True, informative=False, 
                               is_group_model=True, include=('z'), trace_subjs=True, p_outlier = 0.05) 

model.sample(5000, burn=200, dbname=str(label + '.db'), db='pickle') 

model.save(str(label))
```

```python
label = "GLM_HMM_RDK_En" 
model = hddm.HDDMRegressor(humans_rdk_eng, v_reg, group_only_regressors=False, 
                               keep_regressor_trace=True, informative=False, 
                               is_group_model=True, include=('z'), trace_subjs=True, p_outlier = 0.05) 

model.sample(5000, burn=200, dbname=str(label + '.db'), db='pickle') 

model.save(str(label))
```

```python
label = "GLM_HMM_RDK_En_wbias" 
model = hddm.HDDMRegressor(humans_rdk_eng_wbias, v_reg, group_only_regressors=False, 
                               keep_regressor_trace=True, informative=False, 
                               is_group_model=True, include=('z'), trace_subjs=True, p_outlier = 0.05) 

model.sample(5000, burn=200, dbname=str(label + '.db'), db='pickle') 

model.save(str(label))
```
