```python
# Load all modules 
import hddm
import multiprocessing as mp

# Load data and normalise it by the standard deviation

m133_RDK = hddm.load_csv('m133_dots_pp.csv')
m133_RDK["stim"] = m133_RDK["stim"]/24.7

m134_RDK = hddm.load_csv('m134_dots_pp.csv')
m134_RDK["stim"] = m134_RDK["stim"]/12.3

m133_cyl = hddm.load_csv('m133_cyl_pp.csv')
m133_cyl["stim"] = m133_cyl["stim"]/0.026

m134_cyl = hddm.load_csv('m134_cyl_pp.csv')
m134_cyl["stim"] = m134_cyl["stim"]/0.02


# Put together a full combination of paramters since one model will be created for each 

# Initialise and print into terminal number of cores
pool = mp.Pool(mp.cpu_count())
print("Number of processors: ", mp.cpu_count())
Full_list_of_arguments = []

# iteration = 1 
v_reg = {'model': 'v ~ stim', 'link_func': lambda x: x}
included = ('z')

for iteration in range (1, 6):
    print(f"iteration number: {iteration}")

    model1 = hddm.HDDMRegressor(m133_RDK, v_reg, group_only_regressors=False, 
                               keep_regressor_trace=True, informative=False, 
                               is_group_model=True, include=included, trace_subjs=True, p_outlier = 0.05) 

    model1.sample(5000, burn=200, dbname=str("NORMm133_RDK" + '_' + 'z' + '_'+ "5percent_outlier_removal" + 
                                            str(iteration) + '.db'), db='pickle') 
    model1.save(str("m133_RDK" + '_' + 'z' + '_' + "5percent_outlier_removal" + str(iteration)))


    model2 = hddm.HDDMRegressor(m134_RDK, v_reg, group_only_regressors=False, 
                               keep_regressor_trace=True, informative=False, 
                               is_group_model=True, include=included, trace_subjs=True, p_outlier = 0.05) 

    model2.sample(5000, burn=200, dbname=str("NORMm134_RDK" + '_' + 'z' + '_'+ "5percent_outlier_removal" + 
                                            str(iteration) + '.db'), db='pickle') 
    model2.save(str("m134_RDK" + '_' + 'z' + '_' + "5percent_outlier_removal" + str(iteration)))

    model3 = hddm.HDDMRegressor(m133_cyl, v_reg, group_only_regressors=False, 
                               keep_regressor_trace=True, informative=False, 
                               is_group_model=True, include=included, trace_subjs=True, p_outlier = 0.05) 

    model3.sample(5000, burn=200, dbname=str("NORMm133_cyl" + '_' + 'z' + '_'+ "5percent_outlier_removal" + 
                                            str(iteration) + '.db'), db='pickle') 
    model3.save(str("m133_cyl" + '_' + 'z' + '_' + "5percent_outlier_removal" + str(iteration)))


    model4 = hddm.HDDMRegressor(m134_cyl, v_reg, group_only_regressors=False, 
                               keep_regressor_trace=True, informative=False, 
                               is_group_model=True, include=included, trace_subjs=True, p_outlier = 0.05) 

    model4.sample(5000, burn=200, dbname=str("NORMm134_cyl" + '_' + 'z' + '_'+ "5percent_outlier_removal" + 
                                            str(iteration) + '.db'), db='pickle') 
    model4.save(str("m134_cyl" + '_' + 'z' + '_' + "5percent_outlier_removal" + str(iteration)))

```
